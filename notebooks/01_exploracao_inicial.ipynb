```python
# Importações das libs que vamos utilizar no notebook.
from pathlib import Path
import pandas as pd
```

```python
# Criação das variaveis com os caminhos respectivos de cada pasta.
BASE_DIR = Path.cwd().parents[0]
DATA_FOLDER = BASE_DIR / "data"
DATA_RAW = DATA_FOLDER / "raw"
RAW_FILE = DATA_RAW / "respostas_google_forms.csv"
```

```python
# Importando a base de dados para darmos inicio a exploração inicial (EDA)
respostas = pd.read_csv(RAW_FILE)

respostas.head()
```
